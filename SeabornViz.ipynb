Univariant Plots - distplot

Bivarient Plots - jointplot, pairplot

Categorical Plots - countplot, boxplot, swarmplot, heatmap

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
# or via ; tips = pd.read_csv('tips.csv')
tips
```

```python
tips.info()
```

```python
# Plot on one variable on X Axis i.e. Total Bill # Univarince Observations
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
# Joint plots => Bi Variant Graphs
# Distribution of Total Bill over Tip
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg') #prediction line
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
tips.columns
```

```python
#Pair Plots
# Total_Bill, Tip, Size
# 3 Numeric Columns => 9 graphs
# 4 Numeric Columns => 16 graphs
sns.pairplot(tips)
```

```python
# Plot graphs based on categorical columns like sex, smoker, day, time
sns.countplot(x='sex', data=tips)
```

```python
sns.countplot(x='smoker', data=tips)
```

```python
sns.countplot(x='day', data=tips)
```

```python
sns.countplot(x='time', data=tips)
```

```python
# Box plot
sns.boxplot(x='day', y='tip', data=tips)
```

```python
sns.boxplot(x='sex', y='tip', data=tips)
```

```python
sns.boxplot(x='smoker', y='tip', data=tips)
```

```python
sns.boxplot(x='day', y='tip', data=tips, hue='smoker')
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
# heat maps
fp = tips.pivot_table(index='tip', columns='day', values='total_bill')
fp
```

```python
sns.heatmap(fp)
```
